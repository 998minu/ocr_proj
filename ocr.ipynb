```python
from matplotlib import pyplot as pyplt
import numpy as np
plt.figure()
for idx,image in enumerate(X[:3]):
  plt.subplot(1,3,idx+1)
  plt.imshow(np.reshape(image,(28,28)),cmap=plt.cm.gray)
  plt.title('Training:'+y[idx],fontsize=10);
```
